```r
dataset1 <- read.csv("profiling.txt", header = FALSE)
factor1 <- rep(gl(2, 1), each = 72)
factor2 <- rep(gl(3, 3*8), time = 2)
factor3 <- rep(gl(3, 1), time = 2*3*8)
data <- data.frame(
  Response = dataset1,
  Factor1 = factor(factor1),
  Factor2 = factor(factor2),
  Factor3 = factor(factor3)
)
model <- aov(V1 ~ Factor1 * Factor2 * Factor3, data = data)
summary(model)
```
